```python
from textblob import TextBlob
Text1=("I Love this product and it's amazing")
blob1=TextBlob(Text1)
sentiment1=blob1.sentiment.polarity

print("=" * 60)
print("TEXT 1:", Text1)
print(f"Sentiment Score: {sentiment1:.2f}")
if sentiment1 > 0:
    print("Sentiment: PositiveðŸ¤ž")
elif sentiment1 < 0.4:
    print("Sentiment: NuetralðŸ‘")
else:
    print("Sentiment: NegativeðŸ˜¤")
print("=" * 60)
```

```python
reviews = [
    "This phone is absolutely amazing! Best purchase ever!",
    "Battery life is terrible. Dies in 2 hours.",
    "Good phone but a bit expensive for what it offers.",
    "Camera quality is outstanding! Love taking photos with it.",
    "Worst phone I've ever owned. Constantly crashes.",
    "It's okay. Nothing special but does the job.",
    "Incredible value for money! Highly recommend!",
    "Screen broke after one week. Very disappointed.",
    "Fast performance and beautiful design. Very happy!",
    "Customer service was unhelpful when I had issues."
]

print("ðŸ“± PRODUCT REVIEWS TO ANALYZE:")
print("=" * 60)
for i, review in enumerate(reviews, 1):
    print(f"{i}. {review}")
print("=" * 60)
```

```python
import pandas as pd # Import pandas if not already imported

sentiments = []
polarities = []
categories = []

for review in reviews:
    blob = TextBlob(review)
    polarity = blob.sentiment.polarity

    if polarity > 0.1:
        category = "PositiveðŸ˜˜"
    elif polarity < -0.1:
        category = "NegativeðŸ‘¿"
    else:
        category = "NeutralðŸ™ƒ"

    sentiments.append(review)
    polarities.append(polarity)
    categories.append(category)

df = pd.DataFrame({
    'Review': sentiments,
    'Polarity': polarities,
    'Category': categories
})

display(df)
```
