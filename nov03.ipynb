Read the sentences and label them manually as: Positive / Negative / Neutral

```python
#process 1
import pandas as pd

data = {
    "text": [
        "I love reporter and their new reports!",
    "This update was so disappointing.",
    "I think the market is performing great this week.",
    "Not sure how I feel about the latest CBS line.",
    "Absolutely fantastic analysis by the CBS team!"
    ]
}

df = pd.DataFrame(data)
df
```

I love Cbs - Positive This Subject Difficult - Negative I am happy today - positive i hate homework - negative The

```python
#Process 2
import nltk
nltk.download('vader_lexicon')

from nltk.sentiment import SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()

df['score'] = df['text'].apply(lambda x: sia.polarity_scores(x)['compound'])
df['sentiment'] = df['score'].apply(lambda x:
                                    'Positive' if x > 0
                                    else ('Negative' if x < 0
                                    else 'Neutral'))
df
```

```python
#process 3
!pip install afinn
from afinn import Afinn

af = Afinn()

sentences = ["I love this class", "I hate homework", "This is okay"]
for s in sentences:
    print(s, af.score(s))
```

```python
#process 4
!pip install textblob
from textblob import TextBlob

sentences = ["I love this class", "I hate homework", "This is okay"]
for s in sentences:
    blob = TextBlob(s)
    print(s, blob.sentiment)
```

\how VADER differs form Affin and TextBlob?

VADER, AFINN, and TextBlob are all lexicon-based sentiment analysis tools but differ in their focus and capability. VADER is best for social media text as it understands emojis, slang, and emphasis like “!!!” or “not good.” AFINN is simpler, assigning numeric scores to words but lacking context and emoji handling. TextBlob analyzes grammar and context better, giving polarity and subjectivity, but isn’t well-suited for informal or emoji-rich text.

```python
#Let us compare !
from textblob import TextBlob

df['textblob_score'] = df['text'].apply(lambda x: TextBlob(x).sentiment.polarity)
df['textblob_sentiment'] = df['textblob_score'].apply(
    lambda x: 'Positive' if x>0 else ('Negative' if x<0 else 'Neutral')
)
df
```

```python
df['clean_text'] = df['text'].str.lower()
df['clean_text'] = df['clean_text'].str.strip()
df[['text','clean_text']]
```

Why do we convert text to lowercase?

```python
import string
df['no_punct'] = df['clean_text'].str.translate(str.maketrans('', '', string.punctuation))
df[['clean_text','no_punct']]
```

What happens to meaning if punctuation is removed?

```python

import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

stop = set(stopwords.words('english'))
df['no_stopwords'] = df['no_punct'].apply(lambda x: " ".join([word for word in x.split() if word not in stop]))
df[['no_punct','no_stopwords']]
```

Why remove stopwords?

```python
import re

def remove_emojis(text):
    emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"
                           u"\U0001F300-\U0001F5FF"
                           u"\U0001F680-\U0001F6FF"
                           u"\U0001F1E0-\U0001F1FF"
                           "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)

df['no_emoji'] = df['no_stopwords'].apply(remove_emojis)
df[['no_stopwords','no_emoji']]
```
